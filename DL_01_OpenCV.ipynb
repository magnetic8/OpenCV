### 이진(Binary) 이미지 만들기
- 픽셀값 0 ~ 255

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('./images/mouse.jfif', cv2.IMREAD_GRAYSCALE)

# threshold : 한계점
ret, thr1 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)

plt.imshow(thr1, cmap = 'gray')
plt.xticks([])
plt.yticks([])
plt.show()
```

- cv2.threshold(img, threshold_value, value, flag)
    - threshold_value : 픽셀 경계값(0과 255로 나누기 위한 기준)
    - value : 경계값보다 클 때 적용되는 값
    - flag : 적용 방법
        - cv2.THRESH_BINARY : 픽셀값이 경계값보다 크면 value, 아니면 0을 할당
        - cv2.THRESH_BINARY_INV : 픽셀값이 경계값보다 크면 0, 아니면 value 할당
        - cv2.THRESH_TRUNC : 픽셀값이 경계값보다 크면 경계값, 아니면 픽셀값 그대로 할당
        - cv2.THRESH_TOZERO : 픽셀값이 문턱값보다 크면 픽셀값, 아니면 0을 할당
        - cv2.THRESH_TOZERO_INV : 픽셀값이 문턱값보다 크면 0, 아니면 픽셀값을 할당

### 한글 출력하기

```python
import cv2
from PIL import ImageFont, ImageDraw, Image
import numpy as np

img = cv2.imread('./images/mouse.jfif', cv2.IMREAD_COLOR)

# 폰트와 폰트 크기 설정
font = ImageFont.truetype('fonts/gulim.ttc', 30)

# 이미지를 PIL 이미지로 변환
img_pil = Image.fromarray(img)
draw = ImageDraw.Draw(img_pil)

# fill = (blue, green, red, alpha)
draw.text((10,20), '한글출력', font = font, fill =(255, 0, 0, 0))

img = np.array(img_pil)

cv2.imshow('mouse', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
